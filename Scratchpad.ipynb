```python
import cv2
```

```python
img1 = np.zeros(shape=(500,500,3), dtype = np.int32)
img1.shape
```

```python
plt.imshow(img1)
```

```python
cv2.rectangle(img1, pt1=(300,50), pt2=(400,100), color=(0,0,255), thickness=4)
plt.imshow(img1)
```

```python
cv2.circle(img1, center=(50,50), radius=25, color=(255,0,0), thickness=-1)
cv2.line(img1, pt1=(500,500), pt2=(0,0), color=(255,255,255), thickness=2)
font = cv2.FONT_HERSHEY_PLAIN
cv2.putText(img1, text='Chiti Chadhi Pahad Pe', org=(10,400), fontFace = font, fontScale =2, color=(255,255,255), thickness =2)
plt.imshow(img1)
```
